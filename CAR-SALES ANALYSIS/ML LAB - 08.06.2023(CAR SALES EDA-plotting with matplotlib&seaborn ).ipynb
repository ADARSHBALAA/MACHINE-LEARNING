## Subplots using enumerate



#### Let's create list in which there is only the columns which we are going to plot

```python
features=[]
```

```python
for i in data.columns:
     if data[i].dtypes != 'object':
        features.append(i)
```

```python
list(enumerate(features))
```

```python
features
```

```python
 plt.figure(figsize=(20,10))
for i in enumerate(features):
     plt.subplot(4,4,i[0]+1)
     sns.histplot(data=data, x=i[1], color='red')
    
```

### kde plot

```python
plt.figure(figsize=(20,10))
for i in enumerate(features):
     plt.subplot(4,4,i[0]+1)
     sns.kdeplot(data=data, x=i[1], color='red')
    
```

```python
plt.figure(figsize=(13,10))
for i in enumerate(features):
    plt.subplot(4,4,i[0]+1)
    sns.boxplot(data=data, y=i[1], color='red')
```
